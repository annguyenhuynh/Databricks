```python
%sql
SELECT 
  event_time,
  -- Extracts who or what credential initiated the request
  COALESCE(identity_metadata.run_by, user_identity.email) AS recipient_identity,
  -- Safely pulls parameters from the request map
  request_params.share_name AS share_name,
  request_params.table_name AS table_name,
  action_name,
  -- Corrected sub-field based on your system table schema
  response.status_code AS status_code
FROM system.access.audit
WHERE service_name = 'deltaSharing'
  AND action_name IN ('generateTemporaryTableCredentials', 'getShare', 'listTables')
ORDER BY event_time DESC;
```
