```python
from pyspark.sql import SparkSession
import pyspark.sql.functions as F
from pyspark.sql.functions import *
```

```python
spark = SparkSession.builder \
    .appName("practice_3") \
    .config("spark.sql.adaptive.enabled",True) \
    .getOrCreate()
```

```python
%sql
create table if not exists practice.students (
    student_id integer,
    name varchar(20),
    subject varchar(20),
    score integer
)
```

```python
%sql
insert into practice.students values 
(101, 'Alice', 'Math', 85),
(101, 'Alice', 'Chemistry', 85),
(101, 'Alice', 'Physics',78),
(102, 'Bob', 'Math', 90),
(102, 'Bob', 'Physics', 88),
(103, 'Charlie', 'Math', 75),
(103, 'Charlie', 'Chemistry',75),
(104, 'David', 'Math', 88),
(104, 'David', 'Chemistry', 88),
(104, 'David', 'Physics', 95),
(105, 'Eve', 'Chemistry',91)
```

```python
%sql
select distinct
    s1.student_id,
    s1.name,
    s1.subject,
    s1.score
from practice.students s1
join practice.students s2 on s1.student_id = s2.student_id
where s1.score = s2.score and s1.subject <> s2.subject
order by s1.student_id
```

```python
student_report = spark.sql("""
    select distinct
        s1.student_id,
        s1.name,
        s1.subject,
        s1.score
    from practice.students s1
    join practice.students s2 on s1.student_id = s2.student_id
    where s1.score = s2.score and s1.subject <> s2.subject
    order by s1.student_id
""")
```
