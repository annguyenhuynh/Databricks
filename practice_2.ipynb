```python
from pyspark.sql import SparkSession
import pyspark.sql.functions as F
from pyspark.sql.functions import *
```

```python
spark = SparkSession.builder \
    .appName("practice_2") \
    .config("spark.sql.adaptive.enabled",True) \
    .getOrCreate()

spark
```

```python
%sql
create schema if not exists practice;
```

```python
%sql
create table if not exists employees (
    first_name varchar (50),
    last_name varchar(50),
    doj date,
    department varchar(20),
    salary float
);
```

```python
df = spark.table("employees")
```

```python
df = df.withColumn("Name",concat_ws(" ",df.first_name,df.last_name))
```

```python
df = df.withColumn(
    "new_salary",
    when(col("department")=="IT", (col("salary")*1.17).cast("float"))
    .when(col("department")=="Engineering",(col("salary")*1.25).cast("float"))
    .otherwise(col("salary"))
)
```

```python
df.show()
```

```python
df_final = df.select(
    "Name",
    col("doj").alias("DOJ"),
    col("department").alias("Department"),
    col("salary").alias("Current Salary"),
    col("new_salary").alias("Offered Salary")
)
```
