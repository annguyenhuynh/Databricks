```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
```

```python
spark = SparkSession.builder \
    .appName("JSON transformation") \
    .getOrCreate()
```

```python
%sql
select * from gizmobox.silver.orders_json;
```

#### Access elements from JSON object

```python
%sql

select 
  json_value.order_id,
  json_value.order_status,
  json_value.payment_method,
  json_value.total_amount,
  json_value.transaction_timestamp,
  json_value.customer_id,
  json_value.items
from gizmobox.silver.orders_json
```

#### Dedup Array Elements

```python
%sql

select 
  json_value.order_id,
  json_value.order_status,
  json_value.payment_method,
  json_value.total_amount,
  json_value.transaction_timestamp,
  json_value.customer_id,
  array_distinct(json_value.items)
from gizmobox.silver.orders_json
```

```python
%sql

select 
  json_value.order_id,
  json_value.order_status,
  json_value.payment_method,
  json_value.total_amount,
  json_value.transaction_timestamp,
  json_value.customer_id,
  explode(array_distinct(json_value.items)) as item
from gizmobox.silver.orders_json
```

```python
%sql
create or replace temp view tv_orders_exploade
as
select 
  json_value.order_id,
  json_value.order_status,
  json_value.payment_method,
  json_value.total_amount,
  json_value.transaction_timestamp,
  json_value.customer_id,
  explode(array_distinct(json_value.items)) as item
from gizmobox.silver.orders_json
```

```python
df = spark.sql("""
    select 
  order_id,
  order_status,
  payment_method,
  total_amount,
  transaction_timestamp,
  customer_id,
  item.item_id,
  item.name,
  item.price,
  item.quantity,
  item.category,
  item.details.brand,
  item.details.color
from tv_orders_exploade
""")
```

```python
df.write.format("delta").mode("overwrite").saveAsTable("gizmobox.silver.orders")
```
