```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
```

```python
spark = SparkSession.builder \
    .appName("dataProfiling") \
    .config("spark.sql.adpative.enabled",True) \
    .getOrCreate()
```

```python
df = spark.table("v_customers")
```

```python
df.printSchema()
```

```python
df.select([
    sum(col(c).isNull().cast("int")).alias(c) 
    for c in df.columns
]).show()
```

```python
total_rows = df.count()
print(total_rows)
unique_rows = df.dropDuplicates().count()
print(unique_rows)
dup_rows = total_rows - unique_rows

print(f"Duplicate rows: {dup_rows}")
```

```python
%sql 

select count_if(customer_id is null), count_if(email is null),count_if(customer_name is null),count_if(telephone is null)
from gizmobox.bronze.v_customers;
```
